```python
Stock_price_week['公司獨有報酬率％']=np.log(1+(results.resid/100))
# Stock_price_week['公司獨有報酬率％']=results.resid
Stock_price_week['公司獨有報酬率％^3']=Stock_price_week['公司獨有報酬率％']**3
Stock_price_week['公司獨有報酬率％^2']=Stock_price_week['公司獨有報酬率％']**2
```
